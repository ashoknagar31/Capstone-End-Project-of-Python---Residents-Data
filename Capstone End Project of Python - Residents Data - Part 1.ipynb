# Task 2: Data Cleaning

```python
# 1. Replace "_" with NaN in the entire DataFrame
# We begin by addressing missing data. In our dataset, missing values are represented as underscores.
# We'll replace these underscores with NaN (Not-a-Number) values using the replace function to facilitate further data processing.
df.replace('-', np.nan, inplace=True)
df.head(13)
```

```python
# 2. Handling Missing Values:
# We identify and remove two columns, 'Unnamed: 0' and 'Owner's Spouse Name,' as they are not relevant to our analysis. Dropping unnecessary columns helps simplify the dataset.
df.drop(['Unnamed: 0', 'Owner\'s Spouse Name'], axis=1, inplace=True)
df.head(13)
```

```python
# 3. Data Type Conversion:
# We ensure that relevant columns are of the correct data types for analysis. Specifically,
# we convert 'No of Resident' and 'Confirmed Members' columns to numeric data types while handling any conversion errors by setting them to NaN.
df['No of Resident'] = pd.to_numeric(df['No of Resident'], errors='coerce')
df['Confirmed Members'] = pd.to_numeric(df['Confirmed Members'], errors='coerce')
df['Maintenance Amt'] = pd.to_numeric(df['Maintenance Amt'], errors='coerce')
df.head(20)
```

```python
# 4. Impute Missing Values in 'Maintenance Amt' , ‘Flat Vacancy’, 'Donation', ‘No of Resident' , 'Confirmed Members', 'Availability of owner' and 'Origin of Owner'
# (a) To address missing values in the 'Maintenance Amt' column, we iteratively go through the DataFrame.
# For each missing value, we impute it based on the corresponding 'Flat Area (sq.mt)' value.
# We reference a previously calculated dictionary, 'avg_maintenance_amt,' to fill in these values

# After conversion to numeric, check and calculate mean maintenance amount again
avg_maintenance_by_area = df.groupby('Flat Area (sq.mt)')['Maintenance Amt'].mean().dropna()
avg_maintenance_amt = avg_maintenance_by_area.to_dict()
for index, row in df.iterrows():

    if pd.isnull(row['Maintenance Amt']):

        flat_area = row['Flat Area (sq.mt)']

        if flat_area in avg_maintenance_amt:

            df.at[index, 'Maintenance Amt'] = avg_maintenance_amt[flat_area]
df.head(15)
```

```python
# (b) We handle missing values in the 'Flat Vacancy' column by using a conditional approach.
# If 'Availability of owner' is 'Yes,' we assume the flat is owned ('Flat Vacancy' is 'Owned'),
# and if 'Availability of owner' is not 'Yes,' we consider the flat as vacant ('Flat Vacancy' is 'Vacant').
df['Flat Vacancy'].fillna(df.apply(lambda row: 'Owned' if row['Availlability of owner'] == 'Yes' else 'Vacant', axis=1), inplace=True)
df.head(20)
```

```python
# (c) For missing values in the 'Donation', ‘No of Resident' , 'Confirmed Members' column, we impute them with the median value of the respective column.
df['Donation'].fillna(df['Donation'].median(), inplace=True)
df['No of Resident'].fillna(df['No of Resident'].median(), inplace=True)
df['Confirmed Members'].fillna(df['Confirmed Members'].median(), inplace=True)
df.head(15)
```

```python
# (d) Finally, we handle missing values in the 'Availability of owner' and 'Origin of Owner' columns by imputing them with the mode (most frequent value) of their respective columns.
mode_availability = df['Availlability of owner'].mode()[0]
print(mode_availability)
df['Availlability of owner'].fillna(mode_availability, inplace=True)
mode_origin = df['Origin of Owner'].mode()[0]
df['Origin of Owner'].fillna(mode_origin, inplace=True)
df.head(20)
```

# Task 3: Data Preparation and Preprocessing

```python
# 1. Calculate "Outsiders":
# Calculate the difference between the "Confirmed Members" and "No of Resident" columns and store the result in a new column called "Outsiders."
df['Outsiders'] = df['Confirmed Members'] - df['No of Resident']
df.head(10)
```

```python
# 2. Set Values in "Outsiders" Column:
# In this step, you update the values in the "Outsiders" column. If the calculated difference is greater than zero (indicating that there are more confirmed members than residents),
# you set the value to x, indicating the number of outsiders.
# If the difference is less than or equal to zero (indicating that there are no more confirmed members than residents),
# you set the value to 0, indicating that the person is a resident.
df['Outsiders'] = df['Outsiders'].apply(lambda x: x if x > 0 else 0)
df.head(10)
```

```python
# 3. Standardize Columns:
# You convert the values in the  'Origin of Owner', 'Tenant or owner', 'Availability of owner' and 'Flat Vacancy' columns to lowercase to ensure uniform formatting.
# Lastly, you standardize the 'Owner Name' column by converting it to title case to make names consistent and readable
df['Tenant or owner'] = df['Tenant or owner'].str.lower()
df['Availlability of owner'] = df['Availlability of owner'].str.lower()
df['Flat Vacancy'] = df['Flat Vacancy'].str.lower()
df['Origin of Owner'] = df['Origin of Owner'].str.lower()
df['Owner Name'] = df['Owner Name'].str.title()
df.head()
```

```python
# Demographic Overview: Start by providing an overview of the demographic characteristics of the residents.
# Calculate statistics like the average number of residents per flat, the distribution of residents by origin, and any other relevant demographic information.
# 1. Demographic Overview:
# (a) Calculate the average number of residents per flat.

avg_residents_per_flat = round(df['No of Resident'].mean(), 2)

# (b) Determine the distribution of residents by their place of origin.

origin_distribution = df['Origin of Owner'].value_counts()

# (c) Visualize the distribution of residents by origin using a bar plot.

plt.figure(figsize=(10, 6))
sns.barplot(x=origin_distribution.index, y=origin_distribution.values)
plt.title('Distribution of Residents by Origin')
plt.xlabel("Origin of Owner")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
# Flat Information: Analyze the flat information, such as flat area and number of rooms.
# Calculate statistics like the average flat area and the most common number of rooms in the flats.
# 2. Flat Information:
# (a) Calculate the average flat area (in square meters).

average_flat_area = round(df['Flat Area (sq.mt)'].mean(), 2)

# (b) Find the most common number of rooms in flats.

most_common_rooms = df['No of Room'].mode().values[0]

# (c) Visualize the distribution of flat areas using a histogram.

plt.figure(figsize=(8, 5))
sns.histplot(data=df, x = 'Flat Area (sq.mt)', bins=20, kde=True)
plt.title('Flat Area Distribution')
plt.ylabel("Frequency")
plt.show()
```

```python
# Owner and Tenant Information: Explore the data to understand the distribution between flat owners and tenants.
# Calculate the percentage of flat owners and tenants in the housing society.
# 3. Owner and Tenant Information:
# (a) Calculate the percentage of flat owners and tenants in the housing society.

percentage_Owners = (df['Tenant or owner'] == 'owner').mean() * 100
percentage_Tenant = (df['Tenant or owner'] == 'tenant').mean() * 100

# (b) Visualize the distribution of flat ownership using a pie chart.

plt.figure(figsize = (6,6))
plt.pie([percentage_Owners, percentage_Tenant], labels = ['Owner', 'Tenant'], autopct='%1.1f%%')
plt.title('Distribution of Flat Owners and Tenants')
plt.show()
```

```python
# 4. Print Calculated Statistics:
# Print the calculated statistics, including the average number of residents per flat, average flat area, most common number of rooms, and percentages of flat owners and tenants.
print('Average Number of Residents per Flat:',avg_residents_per_flat)
print(f'\nAverage Flat Area: {average_flat_area} sq.mt')
print('\nMost Common Number of Rooms in Flats:',most_common_rooms)
print(f'\nPercentage of Flat Owners: {percentage_Owners}%')
print(f'\nPercentage of Tenants: {percentage_Tenant}%')
```

```python
# Resident Participation Analysis: Calculate the percentage of residents who have confirmed their participation in the upcoming event.
# Visualize resident participation by flat type (owner/tenant) using a boxplot.
# 1. Resident Participation:
# (a) Calculate the percentage of residents' participation by dividing the 'Confirmed Members' by the 'No of Resident' columns and multiplying by 100 to get the percentage.
percentage_participation = (df['Confirmed Members'] / df['No of Resident'])*100
# (b) Visualize the data by creating a box plot that shows resident participation based on flat type (Owner or Tenant).
plt.figure(figsize = (8,5))
sns.boxplot(x = df['Tenant or owner'], y = percentage_participation, palette=['steelblue', 'green'])
plt.title('Resident Participation by Flat Type')
plt.xlabel('Flat Type (Owner/Tenant)')
plt.ylabel('Percentage of Participation')
plt.show()
```

```python
# Donations Analysis: Calculate the average donation amount contributed by residents.
# Calculate the total donation amount collected from residents.
# 2. Donations Analysis:
# (a) Calculate the average donation amount by finding the mean of the 'Donation' column.
average_donation_amount = df['Donation'].mean()
print(f'Average Donation Amount: {average_donation_amount}')
# (b) Calculate the total donation amount by summing the 'Donation' column.
total_donation_amount = df['Donation'].sum()
print(f'Total Donation Amount Collected: {total_donation_amount}')
```
